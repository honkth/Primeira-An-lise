```python
df.drop(columns=['porcentagem de margem bruta', 'renda bruta', 'ID da fatura'], inplace = True)
```

```python
df.rename(columns={'Tempo':'Horário'}, inplace = True)
```

```python
df['Data'] = pd.to_datetime(df['Data'])
```

```python
df['Horário'] = pd.to_datetime(df['Horário'])
```

```python
df['Horário']
```

```python
df.duplicated().sum()
```

```python
df.dtypes
```

```python
meses = {1: 'Janeiro', 2:'Fevereiro', 3: 'Março'}
```

```python
df['Data'] = pd.to_datetime(df['Data']).dt.month.map(meses)
```

```python
df.head()
```

```python
df['Horário'] = df['Horário'].dt.strftime('%H')
```

```python
df.head()
```

```python
time = [] #lista vazia
for i in list(map(int,df['Horário'])): #laço for mapeando e transformando string para inteiro
    if i > 8 and i < 13: # 9 - 12 (manhã)
      time.append('Manhã') # Comando Append vai inserir Manhã quando o critério for verdadeiro na lista time.
    elif i > 12 and i < 19: 
      time.append('Tarde')
    else:
      time.append('Noite')
df['Horário'] = time
```

```python
df.head()
```

```python
Agrupamento = df.groupby('Linha de produto')['Avaliação'].mean().reset_index(drop=False)
```

```python
Agrupamento
```

```python
Agrupando_avaliação = Agrupamento.round(1)
```

```python
Agrupando_avaliação
```

```python
Agrupando_avaliação = Agrupando_avaliação.sort_values(by = 'Avaliação', ascending = False)
```

```python
Agrupando_avaliação
```

```python
Figura = sns.color_palette('gnuplot') #Apenas selecionado as cores do gráfico
Figura.reverse()
```

```python
ax = sns.barplot(y = 'Linha de produto', x = 'Avaliação', data = Agrupando_avaliação, palette = Figura)
plt.rcParams['figure.figsize'] = (18, 10) #Aumentar o tamanho da Figura
ax.set_xlabel('Avaliação dos Clientes')
plt.rcParams.update({"font.size": 18}) #Aumentar a letra
plt.savefig('Linha_x_avaliacao.png', dpi=300)
```
